```python
print('Welcome to ROCK,PAPER,SCISSORS game !')
print('Please enter your move: (r)ock (p)aper (s)cissors')
#設定代號
win_count = 0
loss_count = 0
tie_count = 0
#起始直接為零，利用數字建構while判斷遊戲是否繼續進行
while(win_count == 0):
    move = input('Your move:  ')
    if move =='r':
        print('ROCK versus...')
    elif move=='s':
        print('SCISSORS versus...')
    elif move=='p':
        print('PAPER versus...')
    else:
        print("Invaild.")
    #以上用if來呈現題目要求表示的 使用者出什麼拳～
    if move == 's':
        move = 1
    elif move == 'r':
        move = 2
    elif move == 'p':
        move = 3
    else:
        print("Invaild.")
    #把上面的拳代換成數字因為後面隨機是用數字～
    import random 
    random.randint(1, 3)
    opponent = random.randint(1,3)
    #關鍵來了！會贏嗎？！
    if opponent == 1:
        print('SCISSORS') 
    elif opponent == 2:
        print('ROCK') 
    elif opponent == 3:
        print('PAPER') 
    #讓玩家知道電腦出什麼（輸給電腦也要知道怎麼輸的ＱＱ
    if move == opponent:
        print('It is a tie!')
        tie_count += 1#計算次數，以下count都是這個意思
    elif (move == 1 and opponent == 2):
        print('You lose!')
        loss_count += 1
    elif move==2 and opponent==3:
        print('You lose!')
        loss_count += 1
    elif move==3 and opponent==1:
        print('You lose!')
        loss_count += 1
    elif move==2 and opponent==1:
        print('You win!')
        win_count += 1
        print('You have', tie_count, 'ties and', loss_count, 'losses before your win.')#遊戲結束同時報告次數
    elif move==3 and opponent==2:
        print('You win!')
        win_count += 1
        print('You have', tie_count, 'ties and', loss_count, 'losses before your win.')
    elif move==1 and opponent==3:
        print('You win!')
        win_count += 1
        print('You have', tie_count, 'ties and', loss_count, 'losses before your win.')
    #（我測試時輸了十二次 有夠雖
        
        
```
